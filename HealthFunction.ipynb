## Testing dungeon health

Goal: Track and update the player's health.

### Steps:

1) Create function to update the health.

````python
# We are going to create a variable that represents our health.
health = 45

def updateHealth(providedHealth):
    # This should print the number 45, which is what we set the variable health to.
    print(providedHealth)
````

Let's look at what the heck is going on with the variables above.

1) We directly defined a variable called ` health ` with the integer 45.
2) We defined a variable called ` providedHealth ` in our ` updateHealth ` function - this doesn't make any sense. How will this work?

` health ` stores an actual value.

` providedHealth ` is a temporary value we are using for this function only. Whatever you put in those parenthasis will be a stand-in value and will replace every instance of ` providedHealth ` in the function. If we use this function and put ` health ` in those parenthasis, then ` health ` will be uses instead of ` providedHealth `. The same goes for the value ` 45 `. If we place ` 45 ` in the parenthasis instead of ` providedHealth `, the value ` 45 ` will replace every instance of ` providedHealth `.

Let's try the function ` updateHealth ` without anything in it.

````python
updateHealth()
````

We get an error because it's expecting a value to use in place of ` providedHealth ` and we didn't provide one. Let's provide one.

````python
updateHealth("test")
````

We placed the string ` test ` in the parenthesis, which replaces every instance of ` providedHealth `. Essentially, the function turns into this:

```
def updateHealth("test"):
    print("test")
```

We know this works, but health is a number, so let's use 45 again.

````python
updateHealth(45)
````

Wonderful! We've entered a health value and printed it out. Now, we need to add a number to update the health by.

* Let's assume we can add health as well as remove it. We can probably assume we can use a positive number to add health and a negative number to subtract health.

* We can also assume that there's a potential to be able to add or subtract different amounts of health depending on the situation - this means we need to be able to pass that number in to the function so we can use the same code, just with a different value.

Let's take our original function:

```
def updateHealth(providedHealth):
    print(providedHealth)
```

and turn it into this:

````python
def updateHealth(providedHealth, numberToChangeHealthBy):
    # We know we want to add the number to change health by for the result.
    print(providedHealth + numberToChangeHealthBy)
````

You can see here that we went ahead and added this - let's test it.

````python
updateHealth(45, 15)

# Remember we're using a positive number to ADD health - we can assume the value SHOULD be 60.
````

When using ` 45 ` in place of ` providedHealth ` and ` 15 ` in place of ` numberToChangeHealthBy `,

Our code turned into:

```
def updateHealth(45, 15):
    print(45 + 15)
```

Let's try this with a NEGATIVE number. We know that we can expect the health to go DOWN.

````python
updateHealth(45, -15)

# We should expect 30 because 45 + (-15) is the same as 45 - 15 which is 30.
````

We need to assume now that we're going to end up using this value somewhere. We need to come up with a way to turn this ` 30 ` we came up with when we completed ` 45 - 15 ` above as a number so we can add or subtract with it later. To do this, we can use a return statement.

````python
def updateHealth(providedHealth, numberToChangeHealthBy):
    return providedHealth + numberToChangeHealthBy
````

Here, we defined our function (don't forget to execute that cell so the function exists!) but we used a return instead of a print statement. This means that the value will no longer print out, but it will still calculate to the number. Let's print this out.

````python
print(updateHealth(45, -15))
````

We still expect this to be 30. We know from putting the print INSIDE of the function that the result is 30. Since we removed the print and added a return, the function itself will then turn into a variable that is 30.

This is the equivalent to the function above:
``` print(45 + (-15)) ```

We can ALSO set this as a variable for the same result.

````python
result = updateHealth(45, -15)

# Since we know that updateHealth(45, -15) evaluates to 30, we can expect it to be 30.
````

Let's check it (don't forget to run the cell to create the result variable)

````python
print(result)
````

This returns as 30 like we expected! Good job, you just updated health.


````python
def nestedInsideUpdateHealth():
    return updateHealth(45, -15)
````

````python
# We expect this to also be 30 because we didn't do anything else to it.

print(nestedInsideUpdateHealth())
````
